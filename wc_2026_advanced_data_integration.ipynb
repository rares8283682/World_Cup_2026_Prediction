```python
def target_balance(df, name):
    balance = (
        df["result"]
        .value_counts()
        .rename_axis("result")
        .reset_index(name="count")
    )

    balance["share"] = (balance["count"] / len(df)).round(4)
    balance["dataset"] = name

    return balance
```

```python
def result_rate_by_bins(df, feature, bins=5):
    temp = df[[feature, "result"]].dropna().copy()
    temp["bin"] = pd.qcut(temp[feature], q=bins, duplicates="drop")

    summary = (
        temp
        .groupby("bin")["result"]
        .value_counts(normalize=True)
        .unstack(fill_value=0)
        .round(3)
    )

    return summary
```

```python
def plot_hist(df, column, title=None, xlabel=None, bins=30):
    plt.figure(figsize=(8, 5))
    plt.hist(df[column].dropna(), bins=bins, edgecolor="black")
    plt.title(title or f"Distribution of {column}")
    plt.xlabel(xlabel or column)
    plt.ylabel("Count")
    plt.grid(alpha=0.3)
    plt.show()
```

```python
def plot_bar_counts(df, column, title=None, top_n=None, normalize=False):
    values = df[column].value_counts(normalize=normalize).head(top_n)

    plt.figure(figsize=(10, 5))
    values.sort_values().plot(kind="barh", edgecolor="black")
    plt.title(title or f"Distribution of {column}")
    plt.xlabel("Share" if normalize else "Count")
    plt.ylabel(column)
    plt.grid(axis="x", alpha=0.3)
    plt.show()
```

```python
def plot_scatter(df, x, y, title=None, alpha=0.3):
    plt.figure(figsize=(8, 5))
    plt.scatter(df[x], df[y], alpha=alpha)
    plt.title(title or f"{x} vs {y}")
    plt.xlabel(x)
    plt.ylabel(y)
    plt.grid(alpha=0.3)
    plt.show()
```

```python
def plot_corr_heatmap(df, columns, title):
    corr = df[columns].corr()

    plt.figure(figsize=(10, 8))
    sns.heatmap(corr, cmap="coolwarm", center=0, annot=True, fmt=".2f")
    plt.title(title)
    plt.tight_layout()
    plt.show()
```

### SPI Feature Correlation

The SPI features are highly correlated because they come from the same pre-match SPI model. Home win probability, away win probability, SPI difference, projected score difference and probability difference all measure similar information about team strength and expected match outcome.

For a simple model, it is better to avoid using all SPI features at once. The cleanest SPI feature set is the three SPI probabilities: home win probability, draw probability and away win probability. These already summarize FiveThirtyEight's pre-match prediction.

### SPI Scatter Plot Interpretation

The scatter plots show that the SPI features are strongly related. SPI rating difference, projected score difference and probability difference all describe the same underlying idea: which team is stronger before the match.

The relationship between projected score difference and probability difference is especially strong and nonlinear. This means including all SPI-derived features may add redundancy rather than new information.

For modeling, I will start with a compact SPI feature set using only the SPI probabilities. These probabilities already summarize the pre-match SPI prediction.

```python
def plot_result_rate_by_bins(df, feature, bins=5):
    temp = df[[feature, "result"]].dropna().copy()
    temp["bin"] = pd.qcut(temp[feature], q=bins, duplicates="drop")

    rates = (
        temp
        .groupby("bin", observed=False)["result"]
        .value_counts(normalize=True)
        .unstack(fill_value=0)
    )

    rates.plot(kind="bar", figsize=(10, 5), edgecolor="black")
    plt.title(f"Result Share by {feature} Bin")
    plt.xlabel(feature)
    plt.ylabel("Share")
    plt.xticks(rotation=45, ha="right")
    plt.grid(axis="y", alpha=0.3)
    plt.tight_layout()
    plt.show()

    return rates.round(3)


display(plot_result_rate_by_bins(df_spi_experiment, "spi_diff"))
display(plot_result_rate_by_bins(df_spi_experiment, "spi_prob_diff"))
display(plot_result_rate_by_bins(df_spi_experiment, "spi_proj_score_diff"))
```

```python
from sklearn.model_selection import TimeSeriesSplit

def show_time_series_cv_splits(df, features, name, n_splits=4):
    df_cv = df.sort_values("date").copy()

    X_cv = df_cv[features]
    y_cv = df_cv["result"]

    tscv = TimeSeriesSplit(n_splits=n_splits)

    rows = []

    for fold, (train_idx, val_idx) in enumerate(tscv.split(X_cv), start=1):
        rows.append({
            "dataset": name,
            "fold": fold,
            "train_start": df_cv.iloc[train_idx]["date"].min(),
            "train_end": df_cv.iloc[train_idx]["date"].max(),
            "val_start": df_cv.iloc[val_idx]["date"].min(),
            "val_end": df_cv.iloc[val_idx]["date"].max(),
            "train_rows": len(train_idx),
            "val_rows": len(val_idx),
        })

    return pd.DataFrame(rows)
```

```python
def date_block_cv_splits(df, date_col="date", n_splits=4):
    df = df.sort_values(date_col).copy()

    unique_dates = pd.Series(df[date_col].drop_duplicates().sort_values().values)

    date_folds = np.array_split(unique_dates, n_splits + 1)

    rows = []

    for fold in range(1, n_splits + 1):
        train_dates = pd.concat(date_folds[:fold])
        val_dates = date_folds[fold]

        train_idx = df[df[date_col].isin(train_dates)].index
        val_idx = df[df[date_col].isin(val_dates)].index

        rows.append({
            "fold": fold,
            "train_start": df.loc[train_idx, date_col].min(),
            "train_end": df.loc[train_idx, date_col].max(),
            "val_start": df.loc[val_idx, date_col].min(),
            "val_end": df.loc[val_idx, date_col].max(),
            "train_rows": len(train_idx),
            "val_rows": len(val_idx),
            "same_date_overlap": len(
                set(df.loc[train_idx, date_col]) &
                set(df.loc[val_idx, date_col])
            )
        })

    return pd.DataFrame(rows)
```

```python
def date_block_cv_generator(df, date_col="date", n_splits=4):
    df = df.sort_values(date_col).copy()

    unique_dates = pd.Series(df[date_col].drop_duplicates().sort_values().values)
    date_folds = np.array_split(unique_dates, n_splits + 1)

    for fold in range(1, n_splits + 1):
        train_dates = pd.concat(date_folds[:fold])
        val_dates = date_folds[fold]

        train_idx = df[df[date_col].isin(train_dates)].index
        val_idx = df[df[date_col].isin(val_dates)].index

        yield fold, train_idx, val_idx
```

```python
from sklearn.metrics import (
    accuracy_score,
    balanced_accuracy_score,
    f1_score,
    log_loss,
    classification_report,
    confusion_matrix
)

def evaluate_classifier(model, X, y, name):
    preds = model.predict(X)

    if hasattr(model, "predict_proba"):
        probs = model.predict_proba(X)
        loss = log_loss(y, probs, labels=model.classes_)
    else:
        loss = None

    print(name)
    print("Accuracy:", round(accuracy_score(y, preds), 4))
    print("Balanced accuracy:", round(balanced_accuracy_score(y, preds), 4))
    print("Macro F1:", round(f1_score(y, preds, average="macro"), 4))
    print("Weighted F1:", round(f1_score(y, preds, average="weighted"), 4))

    if loss is not None:
        print("Log loss:", round(loss, 4))

    print("\nClassification report:")
    print(classification_report(y, preds))

    print("Confusion matrix:")
    print(confusion_matrix(y, preds, labels=model.classes_))
```

```python
def predict_with_draw_threshold(model, X, draw_threshold=0.30):
    probs = pd.DataFrame(
        model.predict_proba(X),
        columns=model.classes_,
        index=X.index
    )

    preds = probs.idxmax(axis=1)

    draw_mask = probs["draw"] >= draw_threshold
    preds[draw_mask] = "draw"

    return preds, probs


def evaluate_threshold_model(model, X, y, name, draw_threshold):
    preds, probs = predict_with_draw_threshold(
        model,
        X,
        draw_threshold=draw_threshold
    )

    print(f"{name} | draw_threshold={draw_threshold}")
    print("Accuracy:", round(accuracy_score(y, preds), 4))
    print("Balanced accuracy:", round(balanced_accuracy_score(y, preds), 4))
    print("Macro F1:", round(f1_score(y, preds, average="macro"), 4))
    print("Weighted F1:", round(f1_score(y, preds, average="weighted"), 4))
    print("Log loss:", round(log_loss(y, probs, labels=model.classes_), 4))

    print("\nClassification report:")
    print(classification_report(y, preds))

    print("Confusion matrix:")
    print(confusion_matrix(y, preds, labels=model.classes_))
```

```python
def make_two_stage_targets(df):
    is_draw = (df["result"] == "draw").astype(int)

    non_draw = df[df["result"] != "draw"].copy()
    winner = (non_draw["result"] == "home_win").astype(int)

    return is_draw, non_draw, winner
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

y_train_draw, train_non_draw, y_train_winner = make_two_stage_targets(spi_train)
y_val_draw, val_non_draw, y_val_winner = make_two_stage_targets(spi_val)

draw_model = Pipeline([
    ("scaler", StandardScaler()),
    ("model", LogisticRegression(
        max_iter=1000,
        class_weight="balanced",
        random_state=42
    ))
])

draw_model.fit(X_train_spi_compact, y_train_draw)
```

```python
winner_model = Pipeline([
    ("scaler", StandardScaler()),
    ("model", LogisticRegression(
        max_iter=1000,
        random_state=42
    ))
])

winner_model.fit(
    train_non_draw[features_spi_compact],
    y_train_winner
)
```

```python
def predict_two_stage(draw_model, winner_model, X, draw_threshold=0.50):
    draw_probs = draw_model.predict_proba(X)[:, 1]

    preds = []

    for i, draw_prob in enumerate(draw_probs):
        row = X.iloc[[i]]

        if draw_prob >= draw_threshold:
            preds.append("draw")
        else:
            winner_pred = winner_model.predict(row)[0]
            preds.append("home_win" if winner_pred == 1 else "away_win")

    return pd.Series(preds, index=X.index)
```

```python
two_stage_results = []

for threshold in [0.40, 0.45, 0.50, 0.55, 0.60]:
    preds = predict_two_stage(
        draw_model,
        winner_model,
        X_val_spi_compact,
        draw_threshold=threshold
    )

    two_stage_results.append({
        "draw_threshold": threshold,
        "accuracy": accuracy_score(y_val_spi, preds),
        "balanced_accuracy": balanced_accuracy_score(y_val_spi, preds),
        "macro_f1": f1_score(y_val_spi, preds, average="macro"),
        "weighted_f1": f1_score(y_val_spi, preds, average="weighted"),
        "draw_predictions": (preds == "draw").sum(),
    })

two_stage_results = pd.DataFrame(two_stage_results)
display(two_stage_results.sort_values("macro_f1", ascending=False))
```

```python
best_threshold = two_stage_results.sort_values("macro_f1", ascending=False).iloc[0]["draw_threshold"]

two_stage_preds = predict_two_stage(
    draw_model,
    winner_model,
    X_val_spi_compact,
    draw_threshold=best_threshold
)

print("Best threshold:", best_threshold)
print(classification_report(y_val_spi, two_stage_preds))
print(confusion_matrix(y_val_spi, two_stage_preds, labels=["away_win", "draw", "home_win"]))
```

### SPI Train, Validation and Test Split

The SPI experiment uses a time-based split because match data is chronological. The model trains on 2019-2021, validates on 2022 and tests on 2023. The small number of 2024 rows is kept as an optional holdout.

Because the SPI dataset is smaller than the full historical dataset, TimeSeriesSplit is also used inside the training period. This gives multiple chronological validation folds without leaking future validation or test data into training.
